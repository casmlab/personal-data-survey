```r
# tidyverse and presentation
library('dplyr')
library('ggplot2')
library('stargazer')
library('lme4')
library('ggeffects')
library('margins')
library('ordinal')


# display
options(repr.matrix.max.cols=50, repr.matrix.max.rows=100)
options(dplyr.width = Inf)

# a function for handling figure sizes
fig <- function(width, heigth){
    options(repr.plot.width = width, repr.plot.height = heigth)
    }
```

```r
data_file <- '../data/personal_data_survey_responses.Rdata'
```

```r
load(data_file, verbose=TRUE)
```

Recode variables so we can sum them.

```r
df <- df %>% 
    mutate_at(vars(starts_with('researchers_')), ~ recode(., "Yes" = 1, "No" = 0, "I'm not sure." = 0)) %>%
    mutate_at(vars(starts_with('sm_companies_')), ~ recode(., "Yes" = 1, "No" = 0, "I'm not sure." = 0)) %>%
    mutate_at(vars(starts_with('journalists_')), ~ recode(., "Yes" = 1, "No" = 0, "I'm not sure." = 0)) %>%
    mutate_if(is.factor, as.numeric) %>%
    mutate(mean_sens = rowMeans(select(., starts_with("sensitivity")))) %>%
    mutate(acceptable_researchers = rowSums(select(., starts_with("researchers")))) %>%
    mutate(acceptable_smcs = rowSums(select(., starts_with("sm_companies")))) %>%
    mutate(acceptable_journs = rowSums(select(., starts_with("journalists")))) %>%
    mutate(acceptable = rowSums(select(., starts_with("acceptable"))))
```

# Explore

First we look at the data in general. Summarize the variables we're examining and draw some plots.

```r
df %>% summarise(mean = mean(acceptable_researchers, na.rm = TRUE), sd = sd(acceptable_researchers, na.rm = TRUE))
```

```r
df %>% summarise(mean = mean(acceptable_journs, na.rm = TRUE), sd = sd(acceptable_journs, na.rm = TRUE))
```

```r
df %>% summarise(mean = mean(acceptable_smcs, na.rm = TRUE), sd = sd(acceptable_smcs, na.rm = TRUE))
```

```r
df %>% summarise(mean = mean(acceptable, na.rm = TRUE), sd = sd(acceptable, na.rm = TRUE))
```

```r
df %>% summarise(mean = mean(mean_sens, na.rm = TRUE), sd = sd(mean_sens, na.rm = TRUE))

```

```r
df %>% summarise(across(
    .cols = c('acceptable', 'acceptable_researchers', 'acceptable_smcs', 'acceptable_journs', 'sensitivity_7'), 
    .fns = list(Mean = mean, SD = sd), na.rm = TRUE, 
    .names = "{col}_{fn}"
    ))
```

How do I make that table a 3 column: variable, mean, sd?

```r
plot(df$mean_sens, df$acceptable)
```

```r
lm_agg_accept <- lm(acceptable ~ mean_sens, 
          data = df)
summary(lm_agg_accept)
```

```r
lm_screenname <- lm(acceptable ~ sensitivity_7, data = df)
summary(lm_screenname)
```

```r
lm_users <- lm(cbind(acceptable_researchers, acceptable_journs, acceptable_smcs) ~ sensitivity_7 +
               source + 
               trust_scale + dp_scale + pb_scale +
               man_alone + race + straight_alone + age_brackets + education_level + income_level, data = df)
summary(lm_users)
```
