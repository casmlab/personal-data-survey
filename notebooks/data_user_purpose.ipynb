# What format do we need this data in to run a GLMM?

We have three items that vary:
* data user: 
    1. social media companies, 
    2. academic researchers, 
    3. journalists
* data type: 
    1. social media data (collapsing metadata, text, images, timestamps, location), 
    2. cell phone location data, 
    3. voter file, 
    4. grocery store 
* data use: 
    1. social/behavioral research (langauge change, misinformation, COVID-19, protest, emotion, elections), 
    2. research about the natural world (national parks, disaster),
    3. interventions (bots, ads, coupons, mental health, facial recognition)
    
Data needs to take the form of

| Answer | Data User | Data Type | Data Use | Respondent | Control Variables |
| ---- | ----| ---- | ---- | ---- | ---- | 
|No | sm_companies | social media data | social issue | ghjeut | controls |
|Yes | academic_researchers | social media data | intervention | ghjeut | controls |

Need to create the "data type" and "date use" variables from the questions. We can get the "data user" type from the variable name.

## Do respondents answer the same for all user types?

```r
gm1 <- glmer(is_ok ~ question_desc + (1 | responseid) + (1 | source),
            data = longdf, family = binomial, nAGQ=0, control=glmerControl(optimizer = "nloptwrap"))
summary(gm1)
```

```r
gm1_scales <- update(gm1, . ~ . + trust_scale + dp_scale + pb_scale)
summary(gm1_scales)
```

```r
anova(gm1, gm1_scales)
```

Including scale scores improves the model. Let's try adding demographics.

```r
gm1_scales_demos = update(gm1_scales, . ~ . + man_alone + race + straight_alone + age_brackets + education_level + income_level) 
summary(gm1_scales_demos)
```

```r
BIC(gm1, gm1_scales, gm1_scales_demos)
```

```r
plot_model(gm1_scales_demos)
```

## Separate Data User, Type, and Purpose

Switching to `longdf_complete` so that we have the same sample in each model to to compare performance.

```r
gm2 <- glmer(is_ok ~ data_user + data_type + purpose + (1 | responseid) + (1 | source),
            data = longdf_complete, family = 'binomial', nAGQ=0, control=glmerControl(optimizer = "nloptwrap"))
summary(gm2)
```

```r
gm3 <- glmer(is_ok ~ purpose * data_type * data_user + (1 | responseid) + (1 | source),
            data = longdf_complete, family = 'binomial', nAGQ=0, control=glmerControl(optimizer = "nloptwrap"))
summary(gm3)
```

```r
lrtest(gm2, gm3)
```

Likelihood ratio test says the model with interactions fits better.

```r
gm4 <- glmer(is_ok ~ purpose * data_type * data_user + 
             trust_scale + dp_scale + pb_scale +
             man_alone + race + straight_alone + age_brackets + education_level + income_level + 
             (1 | responseid) + (1 | source),
            data = longdf_complete, family = 'binomial', nAGQ=0, control=glmerControl(optimizer = "nloptwrap"))
summary(gm4)
```

```r
# gm5 <- glmer(is_ok ~ data_user + data_type + purpose + (1 | source),
#             data = longdf, family = 'binomial', nAGQ=0, control=glmerControl(optimizer = "nloptwrap"))
# summary(gm5)
```

```r
lrtest(gm3, gm4)
```

`lrtest` output says `gm4` is a better fit - interactions + controls and demographics

```r
gm6 <- glmer(is_ok ~ data_user * data_type * purpose + 
             trust_scale + dp_scale + pb_scale +
             man_alone + race + straight_alone + age_brackets + education_level + income_level + 
             sn_sensitivity + 
             (1 | responseid) + (1 | source),
            data = longdf_complete, family = 'binomial', nAGQ=0, control=glmerControl(optimizer = "nloptwrap"))
summary(gm6)
```

```r
lrtest(gm4, gm6)
```
